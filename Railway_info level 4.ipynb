# Level 4 : Data Visualization and Reporting 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Railway_info level 4.csv')
df.head()
```

### Task 4.1: Visualization 

```python
# mean of train number for each source station

fig, axes = plt.subplots(nrows=1,ncols=1)
fig.set_size_inches(20, 10)

number_trains_per_station_list=df['Source_Station_Name'].unique() 
number_trains_per_station_avg=df.groupby('Source_Station_Name')['Train_No'].mean() 

plt.bar(number_trains_per_station_list, number_trains_per_station_avg, color='g')
plt.xticks(rotation=90, fontsize = 10)
plt.title('Mean of train number for each source station')
plt.xlabel('Name of Source Station')
plt.ylabel('Mean of Train_number')
plt.show()
```

```python
# number of train for each source station

fig, axes = plt.subplots(nrows=1,ncols=1)
fig.set_size_inches(30, 10)

plt.scatter(df['Source_Station_Name'], df['Train_No'])
plt.xticks(rotation=90, fontsize = 10)
plt.title('Number of train for each source station')
plt.xlabel('Name of Source Station')
plt.ylabel('Number of Train')
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=1)
fig.set_size_inches(12, 6)

plt.bar(df['days'], df['Train_No'])
plt.title('Number of train for days')
plt.xlabel('days')
plt.ylabel('Number of Train')
plt.show()
```
